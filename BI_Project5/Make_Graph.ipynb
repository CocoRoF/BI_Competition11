```python
import os
import pandas as pd

import numpy as np
import networkx as nx
from util_2 import df_processing, graph_add_tuple_weights
```

```python
# 필요한 Data Load

df = pd.read_csv('./Data/2023공모전데이터.csv')
data_path = './result_data'
data_list = os.listdir(data_path)
```

```python
# 산출된 각각의 결과들을 비교하기 위한 total_df 파일 형성

total_df = df_processing(data_list, data_path)
temp_df = total_df[~total_df['item_complex'].str.contains('Complex_Name')]
temp_df.reset_index(inplace=True, drop=True)
```

```python
# 해당 결과들로부터, 동시출현한 발전소들의 쌍을 튜플로 만들어 그래프에 추가함.
# 만약 튜플 쌍이 여러번 등장하면, Edge의 가중치를 상승시킴, 이후 gexf 파일로 저장 후 Gephi프로그램을 이용하여 시각화 함

graph = graph_add_tuple_weights(temp_df)
nx.write_gexf(graph, './test.gexf')
```
