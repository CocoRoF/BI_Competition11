```python
import pandas as pd
import os
import numpy as np
from util_2 import df_processing, sum_make
from util import error_masking, usage_masking
```

```python
# 필요한 Data Load

df = pd.read_csv('./Data/2023공모전데이터.csv')
data_path = './result_data'
data_list = os.listdir(data_path)
best_result_df = pd.read_csv('./result_data/result_546.csv')
```

```python
# 산출된 각각의 결과들을 비교하기 위한 total_df 파일 형성

total_df = df_processing(data_list, data_path)

temp_df = total_df[total_df['item_complex'].str.contains('Complex_Name')].sort_values(by='Money_percent', ascending=False)
temp_df.to_excel('./Data/total_df.xlsx', index=False)
```

```python
# best_result에서 각각 집합의 아이템을 뽑아옴
item_list_1 = best_result_df.item_complex[0].split('[')[1].split(']')[0].replace("'",'').replace(" ", '').split(',')
item_list_2 = best_result_df.item_complex[1].split('[')[1].split(']')[0].replace("'",'').replace(" ", '').split(',')
item_list_3 = best_result_df.item_complex[2].split('[')[1].split(']')[0].replace("'",'').replace(" ", '').split(',')

print(len(item_list_1), len(item_list_2), len(item_list_3))
```

```python
# item_list를 바꿔가면서 대입함. 해당 결과의 각 집합별 nMAE를 출력함
# row = 일시, colume = 시간

temp_array = np.zeros([1, 24])
for num, item_list in enumerate([item_list_1, item_list_2, item_list_3]):
  for i in range(len(df.Date.unique())):
    temp = sum_make(i, item_list, df)[3].reshape(1,24)
    temp_array = np.vstack((temp_array, temp))

  temp_df = pd.DataFrame(temp_array)
  temp_df.to_csv(f'./nMAE/item_{num}.csv')
```
