```python
cntry_list = list(df.Country_Name)
```

```python
temp_df = pd.read_excel('/content/drive/MyDrive/work_space/Project_Files/BI공모전/분석과제 2/글로벌 무역현황_2022년도.xlsx')
```

```python
temp = temp_df[(temp_df['수출국가'].str.contains(search)) & (temp_df['수입국가'].str.contains(search)) & (temp_df['수출 점유율'] >= 0.5) & (temp_df['수입 점유율'] >= 0.5)]
```

```python
temp = temp.drop('품목 HS CODE', axis=1)
temp.reset_index(inplace=True, drop=True)
```

```python
temp
```

```python
total_df = pd.concat([total_df, temp], ignore_index = True)
```
